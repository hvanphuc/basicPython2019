#### 3.1. Các định danh (identifers)
Một định danh (***identifier***) trong ***Python*** là một tên được sử dụng để nhận diện một ***biến, một hàm, một lớp, hoặc một đối tượng.***(Các khái niệm này sẽ được giải thích sau, đừng bận tâm chúng)<br>

#### 5. Toán tử
##### 5.1. Toán tử số học
Toán tử số học trả kết quả của một phép tính, ví dụ:

````python
a = 10
b = 3
str="SHPT"
print ("a+b= ",a+b) # Toán tử cộng
print ("a-b= ",a-b) # Toán tử trừ
print ("a*b= ",a*b) # Toán tử nhân 2 số
print ("str*b=", str*b) # Toán tử nhân 1 str với một số
print ("a/b= ",a/b)  # Toán tử chia, kết quả số thực
print ("a//b= ", a//b) # Toán tử chia, kết quả số nguyên
print ("a**b= ", a**b) # Toán tử lũy thừa
````

##### 5.2. Toán tử quan hệ
Toán tử quan hệ trả kết quả ```True``` hoặc ```False``` của phép so sánh, ví dụ:

````python
a = 10
b = 3
str1 ="SHPT"
str2 ="SHPTSHPT"
print ("a>=b?", a >= b) # Toán tử lớn hơn hoặc bằng
print ("a<=b?", a <= b) # Toán tử bé hơn hoặc bằng
print ("a==b?", a == b) # Toán tử bằng (so sánh giá trị)
print ('a!=b?', a != b) # Toán tử không bằng (so sánh giá trị)
print ('str1 <str2?', str1 <str2 #So sánh theo mã ACSII
````

#### 5.3. Toán tử logic
Toán tử logic trả kết quả `True` hoặc `False` của phép logic (`and`, `or`, `not`), ví dụ:

````python
# Dự đoán các kết quả logic sau:
a = 10
b = 3
c = 5
logic1 = (a>b) and (a>c) or (c>b)
logic2 = not (a>b)
print (logic1)
print (logic2)

# Dự đoán các kết quả logic sau:

print(True and False or True)
print(True or False and False)
````

#### 5.4. Toán tử Membership (Toán tử thành phần)
Toán tử membership trả vế kết quả `True` hoặc `False`, ví dụ:

````python
triplet = "ATG"
seq1 ='ATAAAATAATG'
seq2 ='ATCCCCCCCCG'
lstcodon=["ATG", "ACT", "ATC"]

print (triplet in seq1)
print (triplet not in seq2)
print (triplet in seq1 and triplet in lstcodon)
````

### 1. Kiểu dữ liệu String
String là kiểu dữ liệu rất phổ biến trong python và cũng là đối tượng cần xử lý trong các bài toán bioinformtics.<br>
Một string có thể nhận biết bỡi dấu `'string here'` hoặc  `"string here"` hoặc `"""multiple lines string here"""`<br>
Chuỗi string có thể chứa các kí tự thoát như `\t` biểu thị một kí tự ***tab*** hoặc `\n` biểu thị kí tự ***xuông dòng***.<br>
Mỗi String có thể coi là tập hợp các kí tự bên trong nó, do đó có thể truy cập và sửa đổi các kí tự trong chuỗi String thông qua ***vị trí của kí tự (index)***.<br>Xem các ví dụ sau đây:


````python
#Try cập phần tử của chuỗi
DNAseq='TATACATATACGGGC'

print(DNAseq[0])  #Phần tử đầu tiên của chuỗi là vị trí 0th
print(DNAseq[-1])  #Phần tử cuối cùng của chuỗi là vị trí -1th
print(DNAseq[1:])  #Từ phần từ thứ 1th đến hết chuỗi
print(DNAseq[:-1]) #Từ phần từ thứ cuối đến đầu chuỗi (đảo ngược chuỗi), không bao gồm phần tử cuối
print(DNAseq[1:5]) #Từ phần từ thứ thứ  1th đến vị trí 5th, không bao gồm vị trí 5th
print(DNAseq[1:5:2]) # mỗi 2 kí tự, từ vị trí thứ 1th đến vị trí thứ 5th, không bao gồm vị trí thứ 5
print(DNAseq[:]) #  toàn bộ chuỗi (copy chuỗi)

````

del DNAbase['C']

````python
# Một ví dụ khác làm việc với file

def extractSeqLength(file): #lấy chiều dài của các trình tự từ file mutiFasta
    fasta = open(file, 'r').readlines()
    length=[]
    for line in fasta:
        if not line.startswith('>'):
            line=line.strip('\n')
            length.append(len(line))
    #print('Số lượng trình tự:', len(length))
    #print('Chiều dài lớn nhất:', max(length))
    #print('Chiều dài nhỏ nhất:',min(length))
    return (length)
def extractGCcontent(file): #Tính %GC các trình tự từ file mutiFasta
    fasta = open(file, 'r').readlines()
    GClist=[]
    for line in fasta:
        if not line.startswith('>'):
            line=line.strip('\n').upper()
            GCcontent=(line.count('G')+line.count('C'))/len(line)
            GCcontent=round(GCcontent, 2) # làm tròn 2 chữ số
            GClist.append(GCcontent)
    GClist=sorted(GClist)
    return GClist
def extractHeader(file):
    fasta = open(file, 'r').readlines()
    header=[]
    for line in fasta:
        if line.startswith('>'):
            header.append(line.strip('\n'))
    return header
extractSeqLength('GD63_CDs.fasta')
extractGCcontent('GD63_CDs.fasta')
extractHeader('GD63_CDs.fasta')
````

***Vẽ cấu trúc plasmid dựa trên thông tin từ file Genbank, sử dụng Biopython và các thư viện khác***. <br>
(!_!) Nếu thấy nhức đầu hãy bỏ qua, vì nó không phải là mục tiêu tài liệu này hướng đến

````python
from Bio import SeqIO
from Bio.SeqFeature import SeqFeature, FeatureLocation
from Bio.Graphics import GenomeDiagram
from reportlab.lib import colors
from reportlab.lib.units import cm

````

````python
record=SeqIO.read("plasmidGD63.gb", "genbank")

gd_diagram= GenomeDiagram.Diagram('plasmid GD63')
gd_track_features=gd_diagram.new_track (1,name="Annotated Features")
gd_feature_set=gd_track_features.new_set()

for feature in record.features:
    if feature.type != "gene":
        continue
    if len(gd_feature_set) %2 ==0:
        color=colors.blue
    else:
        color=colors.lightblue
    gd_feature_set.add_feature(feature, sigil="ARROW", color=color,label=True, label_size=14,label_angle=0)
    
gd_diagram.draw(format="linear", pagesize='A4',fragments=4,start=0, end=len(record))
gd_diagram.write("PlasmidGD63_linear.pdf", "PDF")
print(record.id)
print(len(record))
print(gd_feature_set)
# Xem file mới được tạo ra, và tận hưởng! Biopython có vẻ không làm tốt công việc này như một số công cụ khác
````

````python
gd_diagram.draw(format="circular",start=0, end=len(record), pagesize=(20*cm, 20*cm), circle_core=0.5)
gd_diagram.write("PlasmidGD63_circle.pdf", "PDF")
````
